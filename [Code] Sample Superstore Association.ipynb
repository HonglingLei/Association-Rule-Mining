```python
import pandas as pd
import time

# import data
file = "/Users/honglinglei/Desktop/数据科学导论/HW/hw2/Sample - Superstore.xls"
data = pd.read_excel(file, sheet_name = "Orders")

t0=time.time()

data1 = data[["Order ID", "Product ID"]] #挑出有用的两列
data21 = data.groupby('Product ID').size().reset_index() #计算每个product出现的次数（即n_X）
data21.columns = ["XID", "n_X"] #重命名表头
data22 = data.groupby('Product ID').size().reset_index() #计算每个product出现的次数（即n_Y）
data22.columns = ["YID", "n_Y"] #重命名表头
data3 = pd.merge(data1, data1, how = "outer", on = ["Order ID"]) #把含OID和PID的表（两张一样的）类似SQL中outer join起来，按Order ID关联
data3.columns = ["Order ID", "XID", "YID"] #重命名表头
data3 = data3.query("XID != YID") #删掉X和Y相同的行（因为前面outer join会并出X、X这种）
data4 = data3.groupby(['XID','YID']).size().reset_index() #计算XY同时出现的次数（即n_XY）
data4.columns = ["XID", "YID", "n_XY"] #重命名表头
data5 = pd.merge(data4, data21, on = "XID") #把n_XY和n_X合在一张表中
data5 = pd.merge(data5, data22, on = "YID") #在前一张表基础上再加入n_Y
data5["n_Transactions"] = len(set(data1["Order ID"])) #加列：计算n_Transactions（用set去重）
data5["Support"] = data5["n_X"]/data5["n_Transactions"] #加列：计算Support(X)=n(X)/n(Total Transactions)
data5["Confidence"] = data5["n_XY"]/data5["n_X"] #加列：计算Confidence（X->Y)=n(XY)/n(X)
data5["Lift"] = data5["Confidence"]/data5["n_Y"] #加列：计算Lift(X->Y)=Confidence(X->Y)/n(Y)
data5 #final version

print("total execution time:", time.time()-t0)
```
